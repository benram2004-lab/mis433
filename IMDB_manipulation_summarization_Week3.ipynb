```python
# Does the sorting result look right to you? What's the problem?
# answer: the gross is not descending in proper order
```

```python
# Resolve this problem of "Gross" and convert its data type to float
# Hint: You may find this webpage useful:
# https://stackoverflow.com/questions/28986489/how-to-replace-text-in-a-column-of-a-pandas-dataframe

imdb['Gross'] = pd.to_numeric(imdb['Gross'].str.replace(',', '', regex=False), errors='coerce')

print("DataFrame Info after Gross conversion:")
imdb.info()
print("\nDataFrame Head after Gross conversion:")
display(imdb.head())
```

```python
# Next, redo the sorting on Gross

# Select all movies released after (>=) 2010 and with IMDB_Rating>=8.5
# Show their title, released year, Certificate, and gross.
# Sort them in descending order of "Gross"

filtered_and_sorted_movies = imdb.loc[
    (imdb['Released_Year'] >= 2010) & (imdb['IMDB_Rating'] >= 8.5),
    ['Series_Title', 'Released_Year', 'Certificate', 'Gross']
].sort_values(by='Gross', ascending=False)

display(filtered_and_sorted_movies)
```

```python
# Add a new column "Runtime_min" by removing the substring " min" in "Runtime"
# Set its data type as int
# Hint: https://stackoverflow.com/questions/36505847/substring-of-an-entire-column-in-pandas-dataframe

imdb['Runtime_min'] = imdb['Runtime'].str.replace(' min', '').astype(int)

display(imdb[['Series_Title', 'Runtime', 'Runtime_min']].head())
```

```python
# Add a new column "Age_Year" by expression: Current year - Released_Year
import datetime
current_year = datetime.datetime.now().year
imdb['Age_Year'] = current_year - imdb['Released_Year']

# Display the head with the new column
display(imdb[['Series_Title', 'Released_Year', 'Age_Year']].head())
```

```python
# Add a new column "Decade" with values as 1980, 1990, 2000, 2010, 2020, etc.
imdb['Decade'] = (imdb['Released_Year'] // 10) * 10

display(imdb[['Series_Title', 'Released_Year', 'Decade']].head())
```

## Part 2: Data Summarization

```python
# Total "Gross" of all top 1000 movies
print(imdb['Gross'].sum())
```

```python
# Average "No_of_Votes" of all movies
average_votes = imdb['No_of_Votes'].mean()
print(average_votes)
```

```python
# Count movies in each decade (e.g., ..., 1980, 1990, 2000, 2010, 2020)
# Sort decades by the number of movies in descending order

print("\nNumber of movies per decade (sorted by count descending):")
display(imdb['Decade'].value_counts())
```

```python
# Count movies by different directors.
# Show the top 10 directors with the most movies in this list
display(imdb['Director'].value_counts().head(10))
```

```python
# For each director, calculate the number of movies, average IMDB_Rating, and total gross.
# Ranked in descending order of total gross
# Show the top 10 directors with the most gross
display(imdb.groupby('Director').agg(number_of_movies=('Series_Title', 'count'),
average_imdb_rating=('IMDB_Rating', 'mean'),
total_gross=('Gross', 'sum')).sort_values(by='total_gross', ascending=False).head(10))
```

```python
# Group movies by decade and director
# In each group (i.e., for each decade and each director),
# calculate the number of movies and average IMDB rating.
# Sort in descending order of movie count

decade_director_summary = imdb.groupby(['Decade', 'Director']).agg(
    number_of_movies=('Series_Title', 'count'),
    average_imdb_rating=('IMDB_Rating', 'mean')
).sort_values(by='number_of_movies', ascending=False)

display(decade_director_summary.head(10))
```

```python
# Find all directors whose movies appeared in at least five different decades.
# Your result should include: director, decade, and the number of movies in the decade.

#group by director
decade_counts_per_director = imdb.groupby(['Director', 'Decade']).size().reset_index(name='number_of_movies')

# Filter
multi_decade_directors = decade_counts_per_director.groupby('Director').filter(lambda x: x['Decade'].nunique() >= 5)


display(multi_decade_directors.sort_values(by=['Director', 'Decade']))
```

```python
# Question 1: top ten highest grossing movies and the amount of gross

highest_grossing = imdb.sort_values(by='Gross', ascending=False).head(10)
display(highest_grossing[['Series_Title', 'Gross']])
```

```python
# Questions 2: highest grossing movie per decade, #result will include decade, director, and gross

highest_grossing_per_decade = imdb.dropna(subset=['Gross']).loc[imdb.dropna(subset=['Gross']).groupby('Decade')['Gross'].idxmax()]
display(highest_grossing_per_decade[['Decade', 'Series_Title', 'Director', 'Gross']])
```
