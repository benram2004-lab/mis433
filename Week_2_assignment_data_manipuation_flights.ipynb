## Question 4. Adding new columns

Currently dep_time and sched_dep_time are convenient to look at, but hard to compute with because they’re not really continuous numbers.

For example, 759 means 7:59 and 801 means 8:01. Their difference is not 42 but 2 minutes.

```python
# Convert dep_time and sched_dep_time to minutes since midnight
import numpy as np

dep_time_notna = flights['dep_time'].notna()
sched_dep_time_notna = flights['sched_dep_time'].notna()

flights['dep_time_minutes'] = np.nan
flights['sched_dep_time_minutes'] = np.nan

flights.loc[dep_time_notna, 'dep_time_minutes'] = \
    (flights.loc[dep_time_notna, 'dep_time'].astype(int).astype(str).str.zfill(4).str[0:2].astype(float) * 60) + \
    (flights.loc[dep_time_notna, 'dep_time'].astype(int).astype(str).str.zfill(4).str[2:4].astype(float))

flights.loc[sched_dep_time_notna, 'sched_dep_time_minutes'] = \
    (flights.loc[sched_dep_time_notna, 'sched_dep_time'].astype(int).astype(str).str.zfill(4).str[0:2].astype(float) * 60) + \
    (flights.loc[sched_dep_time_notna, 'sched_dep_time'].astype(int).astype(str).str.zfill(4).str[2:4].astype(float))

print(flights[['dep_time', 'dep_time_minutes', 'sched_dep_time', 'sched_dep_time_minutes']].head())
```

```python
# Create a new column of arr_time - dep_time.
import numpy as np

arr_time_notna = flights['arr_time'].notna()
flights['arr_time_minutes'] = np.nan
flights.loc[arr_time_notna, 'arr_time_minutes'] = \
    (flights.loc[arr_time_notna, 'arr_time'].astype(int).astype(str).str.zfill(4).str[0:2].astype(float) * 60) + \
    (flights.loc[arr_time_notna, 'arr_time'].astype(int).astype(str).str.zfill(4).str[2:4].astype(float))

flights['calculated_flight_time'] = flights['arr_time_minutes'] - flights['dep_time_minutes']

# Compare this column with air_time.
print(flights[['air_time', 'calculated_flight_time']].head())
print("\nMean difference (air_time - calculated_flight_time):")
print((flights['air_time'] - flights['calculated_flight_time']).mean())
```

```python
# Compare dep_time, sched_dep_time, and dep_delay. How would you expect those three numbers to be related?
# Try creating a column to calculate dep_delay from dep_time and sched_dep_time (and/or other columns if necessary).
flights['calculated_dep_delay'] = flights['dep_time_minutes'] - flights['sched_dep_time_minutes']

# Test your results.
print(flights[['dep_delay', 'calculated_dep_delay']].head())
print("\nMean difference (dep_delay - calculated_dep_delay):")
print((flights['dep_delay'] - flights['calculated_dep_delay']).mean())
print("\nNumber of mismatches:")
print((flights['dep_delay'] != flights['calculated_dep_delay']).sum())
```

## Question 5. Mixing things together

The following questions may require multiple operations above.

```python
# Find the 20 most delayed flights.
# Display the following: year,month,day,carrier,flight,dep_delay,arr_delay,carrier
# How do you want to handle ties?
most_delayed_flights = flights.sort_values(by='dep_delay', ascending=False).head(20)
print(most_delayed_flights[['year', 'month', 'day', 'carrier', 'flight', 'dep_delay', 'arr_delay']])
```

```python
# Sort all AA flights to find the top 10 fastest (highest speed) flights.
# Display the following: year,month,day,carrier,flight,orig,dest,distance,air_time,speed (miles per hour)

aa_flights = flights[flights['carrier'] == 'AA'].copy()
aa_flights['speed'] = aa_flights['distance'] / (aa_flights['air_time'] / 60)
fastest_aa_flights = aa_flights.sort_values(by='speed', ascending=False).head(10)
print(fastest_aa_flights[['year', 'month', 'day', 'carrier', 'flight', 'origin', 'dest', 'distance', 'air_time', 'speed']])
```

```python
# Find all flights that satisfy the following:
# - From John F. Kennedy Airpot (JFK) or Newark Aiport (EWR) to Seattle-Tacoma Airport (SEA)
# - Carrier is UA, AA, or DL.
# - Dates from 4/1/2013 (inclusive) to 4/3/2013 (inclusive)
# - Scheduled arrival time is before noon.
# - Display the following: year,month,day,carrier,flight,origin,dest,sched_dep_time,sched_arr_time
# - Sort by year, month, day, sched_arr_time

filtered_flights = flights[
    ((flights['origin'] == 'JFK') | (flights['origin'] == 'EWR')) &
    (flights['dest'] == 'SEA') &
    ((flights['carrier'] == 'UA') | (flights['carrier'] == 'AA') | (flights['carrier'] == 'DL')) &
    (flights['month'] == 4) & (flights['day'] >= 1) & (flights['day'] <= 3) &
    (flights['sched_arr_time'] < 1200)
]

sorted_flights = filtered_flights.sort_values(by=['year', 'month', 'day', 'sched_arr_time'])

print(sorted_flights[['year', 'month', 'day', 'carrier', 'flight', 'origin', 'dest', 'sched_dep_time', 'sched_arr_time']])
```
